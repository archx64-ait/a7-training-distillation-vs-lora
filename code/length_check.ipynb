```python
from transformers import BertTokenizer
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('davidson/labeled_data.csv')

# Load tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Tokenize the dataset and get sequence lengths
token_lengths = [len(tokenizer(text)["input_ids"]) for text in df["tweet"]]

# Plot histogram
plt.hist(token_lengths, bins=50, color='blue', alpha=0.7)
plt.xlabel("Token Length")
plt.ylabel("Number of Tweets")
plt.title("Distribution of Token Lengths in Dataset")
plt.show()

# Find max, mean, and 95th percentile length
max_length = max(token_lengths)
mean_length = sum(token_lengths) / len(token_lengths)
percentile_95 = sorted(token_lengths)[int(0.95 * len(token_lengths))]

print(f"Max Length: {max_length}")
print(f"Mean Length: {mean_length}")
print(f"95th Percentile Length: {percentile_95}")
```
