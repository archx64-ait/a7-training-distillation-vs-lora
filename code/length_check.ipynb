```python
from transformers import BertTokenizer
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('davidson/labeled_data.csv')

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

token_lengths = [len(tokenizer(text)['input_ids']) for text in df['tweet']]

plt.hist(token_lengths, bins=50, color='cyan', alpha=0.7)
plt.xlabel('token length')
plt.ylabel('# of tweets')
plt.title('distribution of token length')
plt.show()

max_length = max(token_lengths)
min_length = min(token_lengths)
mean_length = sum(token_lengths) / len(token_lengths)
percentile_95 = sorted(token_lengths)[int(0.95 * len(token_lengths))]

print(f'max length: {max_length}')
print(f'min length: {min_length}')
print(f'mean Length: {mean_length}')
print(f'95th percentile length: {percentile_95}')
```
